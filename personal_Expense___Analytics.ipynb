```python
import pandas as pd

data = {
    "Date": ["2026-06-01","2026-06-02","2026-06-03","2026-06-04","2026-06-05"],
    "Category": ["Food","Travel","Shopping","Food","Entertainment"],
    "Amount": [250,150,500,300,200]
}

df = pd.DataFrame(data)

print(df)

print("\nTotal Expense =", df["Amount"].sum())

print("\nCategory Wise Expense:")
print(df.groupby("Category")["Amount"].sum())
```

```python
import matplotlib.pyplot as plt

category_expense = df.groupby("Category")["Amount"].sum()

category_expense.plot(kind="bar")

plt.title("Category Wise Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")

plt.show()
```

```python
category_expense.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Expense Distribution")
plt.ylabel("")
plt.show()
```

```python
highest = category_expense.idxmax()
print("Highest Expense Category =", highest)
```

```python
average = df["Amount"].mean()
print("Average Expense =", average)
```

```python
lowest = category_expense.idxmin()
print("Lowest Expense Category =", lowest)
```
